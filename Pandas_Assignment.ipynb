1. How do you load a CSV file into a Pandas DataFrame?



Loading a CSV file into a Pandas DataFrame is typically done using the read_csv() function.

```python
import pandas as pd

# 1. Specify the file path
file_path = '/content/drive/MyDrive/ML_Datasets/titanic.csv'

# 2. Load the CSV into a DataFrame
df = pd.read_csv(file_path)

# You can now view the dataset
df
```

In Pandas, you can create a new column in a DataFrame by performing operations on existing columns. Here is a common way using np.where() for conditional assignment

```python
import numpy as np

# Create a column 'status' based on age, whether a teenager or an adult
df["status"] = np.where(df["age"] > 30, "adult", "teenager")
df
```

The groupby() function in Pandas is used to split data into groups based on one or more columns, then perform aggregate operations (like mean, sum, count, etc.) on each group.

```python
result = df.groupby("sex").mean(numeric_only=True)
result
```

```python
import matplotlib.pyplot as plt

# Average salary per department
avg_salary = df.groupby("sex")["age"].mean()

# Bar chart
avg_salary.plot(kind="bar", title="Average age by gender", color="skyblue", grid=True)
plt.ylabel("age")
plt.show()
```
